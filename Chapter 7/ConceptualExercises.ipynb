## Problem 1

$$f(x) = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \beta_4 (x-\xi)^3_+$$

Where $(x-\xi)^3_+ = (x-\xi)^3$ if $x>\xi$, and 0 otherwise.

a) $$f_1(x) = a_1 + b_1 x + c_1 x^2 + d_1 x^3$$

For $x \le \xi$, the term with $\beta_4$ is equal to zero.

$$f(x) = f_1(x)$$

$$\beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 = a_1 + b_1 x + c_1 x^2 + d_1 x^3$$

Then, in order for both functions to be equal, the coefficient of terms with the same exponent in $x$ must be equal:

$$a_1 = \beta_0$$
$$b_1 = \beta_1$$
$$c_1 = \beta_2$$
$$d_1 = \beta_3$$

b) $$f_2(x) = a_2 + b_2 x + c_2 x^2 + d_2 x^3$$

For $x \gt \xi$:

$$f(x) = f_2(x)$$

$$\beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \beta_4 (x-\xi)^3 = a_2 + b_2 x + c_2 x^2 + d_2 x^3$$

The cubic term $(x-\xi)^3$ can be expanded using the formula:

$$(a-b)^3 = a^3 - 3a^2 b + 3ab^2 - b^3$$

Then:

$$\beta_4(x-\xi)^3 = \beta_4x^3 - 3\beta_4x^2\xi + 3\beta_4x\xi^2 - \beta_4\xi^3$$

Putting it all together:

$$\beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \beta_4x^3 - 3\beta_4x^2\xi + 3\beta_4x\xi^2 - \beta_4\xi^3 = a_2 + b_2 x + c_2 x^2 + d_2 x^3$$

The left-hand side can be rearranged, grouping terms with the same power of $x$:

$$\beta_0 - \beta_4\xi^3 + \beta_1 x + 3\beta_4x\xi^2 + \beta_2 x^2 - 3\beta_4x^2\xi + \beta_3 x^3 + \beta_4x^3 = a_2 + b_2 x + c_2 x^2 + d_2 x^3$$

$$\beta_0 - \beta_4\xi^3 + (\beta_1 + 3\beta_4\xi^2)x + (\beta_2 - 3\beta_4\xi)x^2 + (\beta_3 + \beta_4)x^3 = a_2 + b_2 x + c_2 x^2 + d_2 x^3$$

Now we can equate coefficients with the same power of $x$:

$$a_2 = \beta_0 - \beta_4\xi^3$$
$$b_2 = \beta_1 + 3\beta_4\xi^2$$
$$c_2 = \beta_2 - 3\beta_4\xi$$
$$d_2 = \beta_3 + \beta_4$$

c) $$f_1(\xi) = a_1 + b_1 \xi + c_1 \xi^2 + d_1 \xi^3$$

$$\boxed{f_1(\xi) = \beta_0 + \beta_1 \xi + \beta_2 \xi^2 + \beta_3 \xi^3}$$

$$f_2(\xi) = a_2 + b_2 \xi + c_2 \xi^2 + d_2 \xi^3$$

$$f_2(\xi) = \beta_0 - \beta_4\xi^3 + (\beta_1 + 3\beta_4\xi^2) \xi + (\beta_2 - 3\beta_4\xi) \xi^2 + (\beta_3 + \beta_4) \xi^3$$

Expanding terms for $f_2(\xi)$:

$$f_2(\xi) = \beta_0 - \cancel{\beta_4\xi^3} + \beta_1\xi + \cancel{3\beta_4\xi^3} + \beta_2\xi^2 - \cancel{3\beta_4\xi^3} + \beta_3\xi^3 + \cancel{\beta_4\xi^3}$$

After cancelling terms:

$$\boxed{f_2(\xi) = \beta_0 + \beta_1 \xi + \beta_2 \xi^2 + \beta_3 \xi^3}$$

Then, $f(x)$ is continuous at $\xi$:

$$\boxed{f_1(\xi) = f_2(\xi)}$$

d) If $f(x) = a + b x + c x^2 + d x^3$, then the derivative is $f'(x) = b + 2c x + 3d x^2$. Then:

$$f_1'(\xi) = b_1 + 2c_1 \xi + 3d_1 \xi^2$$
$$f_2'(\xi) = b_2 + 2c_2 \xi + 3d_2 \xi^2$$

Rewriting in terms of betas:

$$f_1'(\xi) = \beta_1 + 2\beta_2 \xi + 3\beta_3 \xi^2$$
$$f_2'(\xi) = \beta_1 + 3\beta_4\xi^2 + 2 (\beta_2 - 3\beta_4\xi) \xi + 3 (\beta_3 + \beta_4) \xi^2$$

Expanding terms for $f_2'(\xi)$:

$$f_2'(\xi) = \beta_1 + 3\beta_4\xi^2 + 2 \beta_2 \xi - 6\beta_4\xi^2 + 3 \beta_3 \xi^2 + 3\beta_4 \xi^2$$

Cancelling terms:

$$f_2'(\xi) = \beta_1 + 2 \beta_2 \xi + 3 \beta_3 \xi^2$$

Then:

$$\boxed{f_1'(\xi) = f_2'(\xi)}$$

e) The second derivative is $f''(x) = 2c + 6dx$. Then:

$$f_1''(\xi) = 2c_1 + 6d_1\xi$$
$$f_2''(\xi) = 2c_2 + 6d_2\xi$$

Rewriting in terms of betas:

$$f_1''(\xi) = 2\beta_2 + 6\beta_3\xi$$
$$f_2''(\xi) = 2(\beta_2 - 3\beta_4\xi) + 6(\beta_3 + \beta_4)\xi$$

Expanding terms for $f_2''(\xi)$:

$$f_2''(\xi) = 2\beta_2 - 6\beta_4\xi + 6\beta_3\xi + 6\beta_4\xi$$

Cancelling terms:

$$f_2''(\xi) = 2\beta_2 + 6\beta_3\xi$$

Then:

$$\boxed{f_1''(\xi) = f_2''(\xi)}$$

## Problem 2

a) For $\lambda = \infty$, the RSS term has no effect. The penalty term is the one that matters. For $m=0$, $g^{(0)}(x) = g(x)$. Then, to minimize the formula, the function $g$ must be equal to 0.

b) For $\lambda = \infty$ and $m=1$, $g^{(1)}(x) = g'(x)$. Then, $g'(x) = 0$, which corresponds to a constant function, $g(x)=k$.

c) For $\lambda = \infty$ and $m=2$, $g^{(2)}(x) = g''(x)$. Then, $g''(x) = 0$, which corresponds to a linear function, $g(x)=ax+b$.

d) For $\lambda = \infty$ and $m=3$, $g^{(1)}(x) = g'''(x)$. Then, $g'(x) = 0$, which corresponds to a quadratic function, $g(x)=ax^2+bx+c$.

e) For $\lambda = 0$, the RSS term is the only one that matters. For $m=3$, it actually doesn't matter, since the second term is null. The function is one that perfectly fits the data points (this may overfit).

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression

n = 50
x = np.linspace(0, 10, n)
eps = np.random.uniform(-20, 20, n)
y = 3*x**2 - 22*x + 20 + eps

data = pd.DataFrame({
    'ones': x**0,
    'X': x,
    'X^2': x**2
})

reg = LinearRegression(fit_intercept=True)

# g equal to zero
plt.plot(x, np.zeros(n), c='y', label="g=0")

# Constant
model = reg.fit(data[['ones']], y)
plt.plot(x, model.intercept_ + model.coef_[0]*x, c='r', label="g'=0")

# Linear
model = reg.fit(data[['X']], y)
plt.plot(x, model.intercept_ + model.coef_[0]*x, c='g', label="g''=0")

# Quadratic
model = reg.fit(data[['X', 'X^2']], y)
plt.plot(x, model.intercept_ + model.coef_[0]*x + model.coef_[1]*x**2, c='k', label="g'''=0")

# Degree n polynomial
poly_data = pd.DataFrame({f'X^{i}': x**i for i in range(0,n)})
columns = poly_data.columns
model = reg.fit(poly_data[columns], y)

x_plot = np.linspace(0, 10, 100)
x_for_prediction = [[k**i for i in range(0,n)] for k in x_plot]
y_plot = model.predict(np.array(x_for_prediction))
plt.plot(x_plot, y_plot, c='tomato', label="RSS=0")

plt.scatter(x, y, label='data')
plt.xlim([-3, 13])
plt.legend();
```
