```python
def Power3(x, a):
    return x**a
```

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(1, 10))
y = [Power3(n, 2) for n in x]

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(1,1,1)
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('$y=x^2$')
ax.set_title('Function $y=x^2$')

# Log scale
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
def PlotPower(x, a):
    y = [n**a for n in x]

    fig = plt.figure(figsize=(6,4))
    ax = fig.add_subplot(1,1,1)
    ax.plot(x, y)
    ax.set_xlabel('x')
    ax.set_ylabel(f'$y=x^{a}$')
    ax.set_title(f'Function $y=x^{a}$')
```

```python
import numpy as np

PlotPower(np.arange(-10, 10), 3)
```
