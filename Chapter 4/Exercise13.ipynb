```python
from ISLP import load_data
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Weekly = load_data('Weekly')
```

```python
Weekly.head()
```

```python
Weekly.shape
```

```python
# Some statistics
Weekly.describe()
```

```python
# Correlation table between features
corr = Weekly.corr(numeric_only=True)
corr.style.format(precision=2).background_gradient(cmap='coolwarm')
```

```python
# Correlation using seaborn

f, ax = plt.subplots(figsize=(6, 6))
sns.heatmap(corr,
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            vmin=-1.0, vmax=1.0,
            square=True, ax=ax);
```

```python
# Pair plots
colors = {'Up': 'blue', 'Down': 'red'}
grr = pd.plotting.scatter_matrix(Weekly, c=Weekly['Direction'].map(colors), figsize=(15, 15), marker='o',
                        hist_kwds={'bins': 40}, s=20, alpha=.8)
```
