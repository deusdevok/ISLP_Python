# The numpy normal distribution

`np.random.normal(loc=0.0, scale=1.0, size=None)`

- loc: the mean of the distribution (default = 0)
- scale: standard deviation (default = 1)
- size: number of samples returned (returns a single random variable by default)

```python
# Generate 50 independent random variables (values) from a N(0,1) distribution
# (Normal distribution with mean (loc) 0 and standard deviation (scale) 1)
x = np.random.normal(size=50)
x
```

```python
# y: add an independent random variable with mean 50 and std 1, N(50,1), to each element of x
y = x + np.random.normal(loc=50, scale=1, size=50)
y
```

```python
# Correlation matrix between x and y
# The off-diagonal elements give the correlation between x and y
np.corrcoef(x,y)
```

```python
# Each time a random distribution is called, the results are different
print(np.random.normal(scale=5, size=2))
print(np.random.normal(scale=5, size=2))
```

```python
# To get the same results each time a random variable is created, a random seed can be used
# rng: random number generator
rng = np.random.default_rng(1303)
print(rng.normal(scale=5, size=2))
rng2 = np.random.default_rng(1303)
print(rng2.normal(scale=5, size=2))
```

```python
# Mean, Variance and Standard Deviation 
rng = np.random.default_rng(3)
y = rng.standard_normal(10) # N(0,1) with size=10
np.mean(y), y.mean()
```

```python
np.var(y), y.var(), np.mean((y-y.mean())**2)
```

```python
# np.var() divides by the sample size n by default, rather than n-1
# ddof=0 by default (delta degrees of freedom). The division in the calculation is n-ddof
np.var?
```

```python
np.sqrt(np.var(y)), np.std(y)
```

```python
# Apply np.mean(), np.var(), np.std() to rows and columns of a matrix
# X: 10x3 matrix of N(0,1) random variables
X = rng.standard_normal((10, 3))
X
```

```python
# axis=0: rows (row-major order)
# Compute the mean across rows
X.mean(axis=0)
```

# Indexing data

```python
# Two-dimensional numpy array
A = np.array(np.arange(16)).reshape((4,4))
A
```

```python
A[1,2] # [row, column] (0 indexed)
```

```python
# Select multiple rows at a time
A[[1,3]] # retrieve second and fourth rows
```

```python
# Select the first and third columns
# Pass colon : to the first argument to select all rows
A[:, [0,2]]
```

```python
# Select submatrix made up of the second and fourth rows,
# as well as first and third columns
A[[1,3], [0,2]] # This fails the task, giving a one-dimensional array of length two
```

```python
# The previous code cell is the same as
np.array([A[1,0], A[3,2]])
```

```python
# This fails to extract the submatrix of
# second and fourth rows and first, third and fourth columns
A[[1,3], [0,2,3]]
```

```python
# One way of creating a submatrix is to first subset rows, then columns
A[[1,3]][:,[0,2]]
```

```python
# Another way to do it is with np.ix_
# It creates a mesh object
# https://numpy.org/doc/stable/reference/generated/numpy.ix_.html
idx = np.ix_([1,3],[0,2,3])
A[idx]
```

```python
# Same using slices
# start:stop:step
A[1:4:2,0:3:2]
```

```python
# This is equivalent to using the slice() function
A[slice(1,4,2), slice(0,3,2)]
```

```python
# Boolean indexing
# Vector of 0's (booleans) of length equal to first dim of A
keep_rows = np.zeros(A.shape[0], bool)
keep_rows
```

```python
keep_rows[[1,3]] = True
keep_rows
```

```python
# apply == to two arrays, does it elementwise
np.all(keep_rows == np.array([0,1,0,1])) # np.all check if all elements are True
```

```python
# Using boolean and numbers index different rows (numbers and booleans are treated differently by numpy)
A[np.array([0,1,0,1])]
```

```python
A[keep_rows]
```

```python
# np.ix_(): create mesh with second and fourth rows, first, third and fourth columns
# apply the function to booleans
keep_cols = np.zeros(A.shape[0], bool)
keep_cols[[0,2,3]] = True
idx_bool = np.ix_(keep_rows, keep_cols)
A[idx_bool]
```

```python
# We can mix a list with array of booleans in np.ix_()
idx_mixed = np.ix_([1,3], keep_cols)
A[idx_mixed]
```

# For loops

```python
# Loop over elements of a list and compute their sum
total = 0
for value in [3,2,19]:
    total += value
print('Total is: {0}'.format(total))
```

```python
# Nesting for loops
total = 0
for value in [2,3,19]:
    for weight in [3,2,1]:
        total += value * weight
print('Total is: {0}'.format(total))
```

```python
# Using zip() function
total = 0
for value, weight in zip([2,3,19],
                        [0.2,0.3,0.5]):
    total += weight * value
print('Weighted average is: {0}'.format(total))
```

## Exercise 3 curves

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10, 20)

bias = 1.8*np.exp(-0.4*x)
variance = 0.032*np.exp(0.4*x) - 0.032
MSE_test = bias + variance + 1

a,b,c,d = np.polyfit(np.array([0, 3, 7, 10]), 
                     np.array([2.2, 1, 0.5, 0]), 
                     3)
MSE_train = a*x**3 + b*x**2 + c*x + d

Bayes = np.ones(20)

plt.figure(figsize=(8, 6))

plt.plot(x, bias, label='Bias')
plt.plot(x, variance, label='Variance')
plt.plot(x, MSE_test, label='MSE Test')
plt.plot(x, MSE_train, label='MSE Train')
plt.plot(x, Bayes, label='Bayes (Irreducible)')

#plt.yticks(np.arange(0, 12, step=1.0))
plt.xlabel('Model Flexibility')
plt.ylabel('Mean Squared Error')
plt.title('Sketch Of Error Curves')
plt.legend()
```
