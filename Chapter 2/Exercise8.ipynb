```python
import pandas as pd
```

```python
college = pd.read_csv('College.csv')
```

```python
college
```

```python
# Make the first column (index 0) the index of the dataframe
college2 = pd.read_csv('College.csv', index_col=0)
college2
```

```python
# Rename the first column. The axis=1 is to specify the renaming is done at the column level
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3
```

```python
# Set the index name for future reference
college3 = college3.set_index('College')
```

```python
college3
```

```python
college = college3
```

```python
# Numerical summary of the variables in the data set
college.describe()
```

```python
# Scatterplot matrix of some columns (use semicolon ; to ignore object outputs)
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll', 'Accept']]);
```

```python
#college.plot.scatter('Accept', 'Enroll')
college.plot.scatter('Accept', 'Apps')
college.loc[college.Apps>40000]
```

```python
college.boxplot(column='Outstate', by='Private')
```

```python
college['Elite'] = pd.cut(
                        college['Top10perc'],
                        [0, 50, 100],
                        labels=['No', 'Yes'])
```

```python
college['Elite']
```

```python
college['Elite'].value_counts()
```

```python
college.boxplot('Outstate', by='Elite')
```
